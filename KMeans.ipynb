```python
import pandas as pd
```

```python
#Create dataframes for each csv file
customers = pd.read_csv('customer_profiles.csv')
details = pd.read_csv('ad_details.csv')
events = pd.read_csv('ad_events.csv')
```

```python
customers.head()
```

```python
#convert custoemrs gender and ever_married to numeric values
for i in range(len(customers)):
    if (customers.iloc[i]['ever_married']) == "No":
        customers.at[i, 'ever_married'] = 0
    else:
        customers.at[i, 'ever_married'] = 1
        
for i in range(len(customers)):
    if (customers.iloc[i]['gender']) == "F":
        customers.at[i, 'gender'] = 0
    else:
        customers.at[i, 'gender'] = 1
```

```python
#convert state to numeric value
num = 0
states_to_num = {}
for state in customers['home_state'].unique():
    states_to_num[state] = num
    num += 1
    
for i in range(len(customers)):
    state = customers.iloc[i]['home_state']
    customers.at[i, 'home_state'] = states_to_num[state]
        
customers
```
